```python
# Freedman-Diaconis rule to determine the optimal bin size for the Price column
q1, q3 = np.percentile(df['Price'], [25, 75])
iqr = q3 - q1
bin_width = 2 * iqr / (len(df['Price'])**(1/3))
n_bins = int((df['Price'].max() - df['Price'].min()) / bin_width)

# Create histogram
plt.hist(df['Price'], bins=n_bins)

# Add labels and title
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Distribution of Price')

# Show plot
plt.show()
```

# Cohort Analysis

```python
df
```

```python
# create invoice month
import datetime as dt
# function for month
def get_month(x):
    return dt.datetime(x.year, x.month, 1) #first day in a month
# apply function
df['InvoiceMonth'] = df['InvoiceDate'].apply(get_month) # this is the month of a certain invoice
```

```python
# that is the month when the user first got an invoice
df['CohortMonth'] = df.groupby('Customer ID')['InvoiceMonth'].transform('min')
```

```python
# create a date element function to get a series for subtraction
def get_date_elements(df, column):
    month = df[column].dt.month
    year = df[column].dt.year
    return month, year 
```

```python
# get date elements for our cohort and invoice columns
Invoice_month,Invoice_year =  get_date_elements(df,'InvoiceMonth')
Cohort_month,Cohort_year =  get_date_elements(df,'CohortMonth')
```

```python
#create a cohort index 
year_diff = Invoice_year -Cohort_year
month_diff = Invoice_month - Cohort_month
df['CohortIndex'] = year_diff*12 + month_diff+1
```

```python
#count the customer ID by grouping by Cohort Month  and Cohort Index 
cohort_data = df.groupby(['CohortMonth','CohortIndex'])['Customer ID'].apply(pd.Series.nunique).reset_index()
```

```python
cohort_data
```

```python
# create a pivot table 
cohort_table = cohort_data.pivot(index='CohortMonth', columns=['CohortIndex'],values='Customer ID')
cohort_table
```

```python
plt.figure(figsize=(30,15))
sns.heatmap(cohort_table,annot=True,cmap='Blues')
```

```python
#cohort table for percentage
new_cohort_percentage = cohort_table.divide(cohort_table.iloc[:,0],axis=0).round(3)
new_cohort_percentage
```

```python
# get rid of the long names
new_cohort_percentage.index = new_cohort_percentage.index.strftime('%Y-%m')
```

```python
# assume retention_table is the DataFrame containing retention rates
plt.figure(figsize=(21,10))
sns.heatmap(new_cohort_percentage, annot=True, cmap='RdYlGn')
plt.title('Cohort Analysis: User Retention')
plt.yticks(rotation=0)
plt.show()
```

```python
#create a percentages visual
plt.figure(figsize=(21,10))
sns.heatmap(new_cohort_percentage,annot=True,fmt='.0%')
```

```python
# Calculate Average Order Value (AOV)
cohort_aov = df.groupby(['CohortMonth', 'CohortIndex'])['Revenue'].mean()
cohort_aov
```

```python
# Calculate Customer Lifetime Value (CLV)
cohort_clv = df.groupby(['CohortMonth', 'CohortIndex'])['Revenue'].sum()
cohort_clv
```

```python
# Calculate Repeat Purchase Rate (RPR)
repeat_customers = df[df['CohortIndex'] > 1]['Customer ID'].nunique()
total_customers = df[df['CohortIndex'] == 1]['Customer ID'].nunique()
rpr = repeat_customers / total_customers
rpr
```

```python
# Calculate Time to Repeat Purchase
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['NextPurchaseDate'] = df.groupby('Customer ID')['InvoiceDate'].shift(-1)
df['TimeToRepeat'] = df['NextPurchaseDate'] - df['InvoiceDate']
avg_time_to_repeat = df[df['CohortIndex'] == 1]['TimeToRepeat'].mean()
```

# Segmentation

```python
# frequency
frequency = df.groupby('Customer ID')['Invoice'].count()
frequency = frequency.reset_index()
frequency
```

```python
recency['Diff'] = recency['Diff'].dt.days // 30# get the days only without time
```

```python
# recency
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate']) # change dateformat
df['Diff'] = max(df['InvoiceDate'])-df['InvoiceDate'] # get the range of first and last
recency1 = df.groupby('Customer ID')['Diff'].min() # find the last invoice date for each customer
recency1 = recency1.reset_index()
recency1['Diff'] = recency1['Diff'].dt.days // 30
recency1
```

```python
# monetary
df['Revenue'] = df['Quantity'] * df['Price']
monetary = df.groupby('Customer ID')['Revenue'].sum()
monetary = monetary.reset_index()
monetary
```

```python
# rfm
rfm = pd.merge(recency, frequency, on='Customer ID', how='inner')
rfm = pd.merge(rfm, monetary, on='Customer ID', how='inner')
rfm.columns = ['Customer ID', 'Recency', 'Frequency', 'Monetary']
rfm
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
K = range(2, 9)
inertia = []

for k in K:
    print("Training a K-Means model with {} clusters! ".format(k))
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(rfm)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
import pickle
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(rfm)

    silhouette.append(silhouette_score(rfm, kmeans.predict(rfm)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Silhouette Method showing the optimal k')
```

```python
# Predict cluster labels for the input data using KMeans algorithm
rfm.drop('labels', axis=1, inplace=True)

labels = kmeans.predict(rfm)

# Extract unique cluster labels from the predicted labels
clusters = np.unique(labels)

# Calculate the cluster centers using KMeans algorithm
centers = kmeans.cluster_centers_

# Loop over the unique cluster labels
for cluster in clusters:
    # Extract the indices of the data points belonging to the current cluster label
    row_ix = np.where(labels == cluster)
    
    # Create a scatter plot of the data points belonging to the current cluster label
    plt.scatter(rfm.to_numpy()[row_ix, 1], rfm.to_numpy()[row_ix, 3], label=f'Cluster {cluster+1}')

# Add the cluster centers to the scatter plot
plt.scatter(centers[:, 1], centers[:, 3], s=200, marker='*', c='black', label='Cluster Centers')

# Add axis labels and title to the plot
plt.xlabel('Recency')
plt.ylabel('Revenue')
plt.title('KMeans Clustering')

# Add a legend to the plot
plt.legend()

# Display the scatter plot
plt.show()
```
